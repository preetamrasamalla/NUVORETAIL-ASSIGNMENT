## Question 1

```python
import os, json
import pandas as pd

path_to_json ='Json_dump/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
print(json_files)  # for me this prints ['foo.json']
```

```python
df=pd.DataFrame(columns=['ASIN','MRP','brandName','productName','browseNode'])
```

```python
ASIN=[]
MRP=[]
brandname=[]
productname=[]
browsenode=[]
for i in json_files:
    with open('C:\\Users\\MSI\\OneDrive - National Institute of Technology Warangal\\Desktop\\SCRIPTS\\Projects\\intern_annaya\\assignment\\Json_dump\\'  +i) as file:
        data=json.load(file)
        if  'attributes' in data.keys() :
            ASIN.append(i.strip('.json'))
            MRP.append(data['attributes']['list_price'][0]['value_with_tax'])
            brandname.append(data['attributes']['brand'][0]['value'])
            productname.append(data['attributes']['item_name'][0]['value'])
            browsenode.append(data['attributes']['recommended_browse_nodes'][0]['value'])
            
        else:
            ASIN.append(i.strip('.json'))
            MRP.append('NA')
            brandname.append('NA')
            productname.append('NA')
            browsenode.append('NA')
           
```

```python
df['ASIN']=pd.DataFrame(ASIN)
df['MRP']=pd.DataFrame(MRP)
df['brandName']=pd.DataFrame(brandname)
df['productName']=pd.DataFrame(productname)
df['browseNode']=pd.DataFrame(browsenode)
```

```python
df.to_excel("Nuvoretail.xlsx",index=False)
```
